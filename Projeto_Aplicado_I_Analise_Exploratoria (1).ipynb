```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregar o dataset
df = pd.read_csv('C:\\Program Files\\Jupyter\\Netflix Userbase.csv')

# Exibir as primeiras linhas do DataFrame
df.head()
```

```python
# Exibindo as informações do dataset
df.info()
```

```python
#Exibindo o tamanho do dataset
df.shape
```

```python
# Exibir informações sobre valores ausentes
df.isnull().sum()
```

```python
# Análise de Receita Mensal
plt.figure(figsize=(10, 6))
sns.histplot(df['Monthly Revenue'], kde=True)
plt.title('Distribuição da Receita Mensal')
plt.xlabel('Receita Mensal')
plt.ylabel('Frequência')
plt.grid()
plt.show()
```

```python
# Análise do Tipo de Assinatura
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Subscription Type',
              hue='Subscription Type', legend=False)
plt.title('Distribuição de Usuários por Tipo de Assinatura')
plt.xlabel('Tipo de Assinatura')
plt.ylabel('Número de Usuários')
plt.grid()
plt.show()
```

```python
# Análise da Receita Mensal por Tipo de Assinatura
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Subscription Type', y='Monthly Revenue',
            hue='Subscription Type', legend=False)
plt.title('Receita Mensal por Tipo de Assinatura')
plt.xlabel('Tipo de Assinatura')
plt.ylabel('Receita Mensal')
plt.grid()
plt.show()
```

```python
# Análise da Idade dos Usuários
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True)
plt.title('Distribuição da Idade dos Usuários')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.grid()
plt.show()
```

```python
# Análise da Distribuição dos Usuários por Gênero
plt.figure(figsize = (10,6))
plt.title("Distribuição por Gênero")
plt.xlabel('Gênero')
plt.ylabel('Número de Usuários')

sns.countplot(data = df, x = "Gender");
plt.show()

print("A quantidade de homens na base é: {}.".format(df["Gender"].value_counts()[1]))
print("A quantidade de mulheres na base é: {}.".format(df["Gender"].value_counts()[0]))
```

```python
# Análise da Distribuição dos Usuários por Gênero e Tipo de Dispositivo
plt.figure(figsize = (10,6))
plt.title('Distribuição de Usuários por Gênero e Tipo de Dispositivo')
plt.xlabel('Tipo de Dispositivo')
plt.ylabel('Número de Usuários')
sns.countplot(data = df, x = "Device", hue = "Gender");
```

```python
# Análise do Tipo de Dispositivo
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Device', hue='Device', legend=False)
plt.title('Distribuição de Usuários por Tipo de Dispositivo')
plt.xlabel('Tipo de Dispositivo')
plt.ylabel('Número de Usuários')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Distribuição por Tipo de Dispositivo
devices = df["Device"].value_counts().reset_index()
devices

plt.figure(figsize = (12,6))

plt.title("Tipo de Dispositivo")

names = devices["Device"]

plt.pie(devices["count"], labels = names, autopct = "%1.1f%%",
        wedgeprops = {"edgecolor" : "black"}, shadow = True);
```

```python
# Análise da Receita Mensal por País
sns.set_style("darkgrid")

grouped_df = df.groupby('Country')['Monthly Revenue'].sum().reset_index()

plt.figure(figsize = (16,7))
plt.title("Receita Mensal por País")
plt.xlabel('País')
plt.ylabel('Receita Mensal')
sns.barplot(x = grouped_df.Country, y = grouped_df["Monthly Revenue"]);
```

```python
# Análise da Receita Mensal por País
if 'Country' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='Country', y='Monthly Revenue',
                hue='Country', legend=False)
    plt.title('Receita Mensal por País')
    plt.xlabel('País')
    plt.ylabel('Receita Mensal')
    plt.xticks(rotation=45)
    plt.grid()
    plt.show()
```
